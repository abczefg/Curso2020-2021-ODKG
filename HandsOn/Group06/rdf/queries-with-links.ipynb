```
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
from google.colab import drive
drive.mount('/gdrive', force_remount=True)
g = Graph()
#g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
#g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse("/gdrive/My Drive/odkg/output-reduced-with-links.nt",format="nt")
```

```
from rdflib.plugins.sparql import prepareQuery
ns = Namespace("http://publicprocurement.odkg.es/group06/ontology/ppg6#")
owl = Namespace("http://www.w3.org/2002/07/owl#")

q1 = prepareQuery('''
  SELECT DISTINCT
    ?publicOrg ?wikidataLink 
  WHERE { 
    ?publicOrg owl:sameAs ?wikidataLink 
  }
  ''',
  initNs = { "ns" : ns, "RDFS": RDFS, "owl" : owl}
)

for r in g.query(q1):
  print(r)
```

```
from rdflib.plugins.sparql import prepareQuery
ns = Namespace("http://publicprocurement.odkg.es/group06/ontology/ppg6#")
owl = Namespace("http://www.w3.org/2002/07/owl#")
wikidata = Namespace("http://www.wikidata.org/wiki/")

q1 = prepareQuery('''
  SELECT DISTINCT
    ?contract
  WHERE { 
    ?contract ns:contractingParty ?publicOrg .
    ?publicOrg owl:sameAs <https://www.wikidata.org/entity/Q20105071> 
  }
  ''',
  initNs = { "ns" : ns, "RDFS": RDFS, "owl" : owl, "wikidata" : wikidata}
)

for r in g.query(q1):
  print(r)
```
