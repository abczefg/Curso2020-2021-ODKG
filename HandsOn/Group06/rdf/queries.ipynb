```
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
from google.colab import drive
drive.mount('/gdrive', force_remount=True)
g = Graph()
#g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
#g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse("/gdrive/My Drive/odkg/output-reduced.nt",format="nt")
```

```
from rdflib.plugins.sparql import prepareQuery
ns = Namespace("http://publicprocurement.odkg.es/group06/ontology/ppg6#")
#SPARQL
q1 = prepareQuery('''
  SELECT DISTINCT
    ?x
  WHERE { 
    ?x a ns:Contract 
  }
  ''',
  initNs = { "ns" : ns, "RDFS": RDFS}
)

for r in g.query(q1):
  print(r)
```

```
ns = Namespace("http://publicprocurement.odkg.es/group06/ontology/ppg6#")
#SPARQL
q1 = prepareQuery('''
  SELECT DISTINCT
    ?x ?date
  WHERE { 
    ?x a ns:Contract .
    ?x ns:awardDate ?date .
    FILTER (?date > "2020-06-01"^^xsd:dateTime) 
  }
  ''',
  initNs = { "ns" : ns, "RDFS": RDFS}
)

for r in g.query(q1):
  print(r)
```

```
ns = Namespace("http://publicprocurement.odkg.es/group06/ontology/ppg6#")
#SPARQL
q1 = prepareQuery('''
  SELECT DISTINCT
    ?publicName ?companyName
  WHERE { 
    ?contract a ns:Contract .
    ?contract ns:contractingParty ?publicOrg .
    ?publicOrg ns:name ?publicName .
    ?contract ns:contractor ?company .
    ?company ns:name ?companyName .
  }
  ''',
  initNs = { "ns" : ns, "RDFS": RDFS}
)

for r in g.query(q1):
  print(r)

```
