Read the RDF file

```
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
g = Graph()
g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse(github_storage+"/resources/example6.rdf", format="xml")
```

**TASK 7.1: List all subclasses of "Person" with RDFLib and SPARQL**

```
ns = Namespace("http://somewhere#")
from rdflib.plugins.sparql import prepareQuery

#RDFLib
def getSubClassesOf(x):
    for s,p,o in g.triples((None, RDFS.subClassOf, x)):
      print(s)
      getSubClassesOf(s)

getSubClassesOf(ns.Person)

#SPARQL
q1 = prepareQuery('''
  SELECT 
    ?SubClass
  WHERE { 
    ?SubClass rdfs:subClassOf/rdfs:subClassOf* ns:Person. 
  }
  ''',
  initNs = { "ns": ns}
)

for r in g.query(q1):
  print(r)
```

**TASK 7.2: List all individuals of "Person" with RDFLib and SPARQL (remember the subClasses)**


```
#RDFLib
def getIndividualsOf(x):
    for s,p,o in g.triples((None, RDF.type, x)): # Instances of x
      print(s)
    for s,p,o in g.triples((None, RDFS.subClassOf, x)): # Instances of subclasses of x
      getIndividualsOf(s)

getIndividualsOf(ns.Person)

#SPARQL
q1 = prepareQuery('''
  SELECT 
    ?ind
  WHERE { 
    ?ind rdf:type/rdfs:subClassOf* ns:Person. 
  }
  ''',
  initNs = { "ns": ns}
)

for r in g.query(q1):
  print(r)
```

**TASK 7.3: List all individuals of "Person" and all their properties including their class with RDFLib and SPARQL**


```
#RDFLib
def getIndividualsAndTheirPropertiesOf(x):
    for s,p,o in g.triples((None, RDF.type, x)): # Instances of x
      for sub,prop,obj in g.triples((s, None, None)): # Properties of x
        print(sub, prop)
    for s,p,o in g.triples((None, RDFS.subClassOf, x)): # Instances of subclasses of x
      getIndividualsAndTheirPropertiesOf(s)

getIndividualsAndTheirPropertiesOf(ns.Person)

#SPARQL
q1 = prepareQuery('''
  SELECT 
    ?ind ?prop
  WHERE { 
    ?ind rdf:type/rdfs:subClassOf* ns:Person.
    ?ind ?prop ?x 
  }
  ''',
  initNs = { "ns": ns}
)

for r in g.query(q1):
  print(r)
```
